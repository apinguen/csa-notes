Java Basics: Variables and Expressions
- Variable names are called symbols
- Data is assigned to a memory location using the equal sign (=)
- Equality: ==
- Datatype declaration \<DataType>\<SymbolName>;
  - int
    - max 32 bits to store value and the sign
    - 2^31 - 1
    - Primitive - exact value stored in memory
  - double
    - max 64 bits
    - 2^63 - 1
    - decimal points
    - approximate representations
  - boolean
  - String
  - float

Casting
- Smaller to larger -> automatically done
- Larger to smaller -> must be declared in the instruction
```
int a = 5;
double b = 10.0;

a = (int) b; // 10.0 is truncated
b = a; // 5 is automatically converted to double

double number;
number = 2026; // 2026

int number;
number = 2.71; // error

int number;
number = (int) 456.789 // 456
```

De Morgan's Laws (Distributing ! to each term)
If not a and b are true, then one or both are false
```
!(a && b) == (!a || !b) == true
```
If neither a nor b are true, both are false
```
!(a || b) == (!a && !b) == true
```
