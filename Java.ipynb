Java Basics: Variables and Expressions
- Variable names are called symbols
- Data is assigned to a memory location using the equal sign (=)
- Equality: ==
- Datatype declaration < DataType >< SymbolName >;
  - int
    - max 32 bits to store value and the sign
    - 2^31 - 1
    - Primitive - exact value stored in memory
  - double
    - max 64 bits
    - 2^63 - 1
    - decimal points
    - approximate representations
  - boolean
  - String
  - float


---


Literally
- Numerical literal - the numbers you see are the value
```
  a = 10; // 10 is literally the value 10
  a = count; // count is not a literal
```

---

Casting
- Smaller to larger -> automatically done
- Larger to smaller -> must be declared in the instruction
```
int a = 5;
double b = 10.0;

a = (int) b; // 10.0 is truncated
b = a; // 5 is automatically converted to double

double number;
number = 2026; // 2026

int number;
number = 2.71; // error

int number;
number = (int) 456.789 // 456
```
Arithmetic Operations
```
// Declaration
int a;
int b;
int c;
// Initialization
a = 4;
b = 5;
c = a + b; // Same type
System.out.println(c);
> 9
```
Arithmetic with Doubles
```
int c = 7.2 + 5.4;
System.out.println(c);
> 12.60000000000001
System.out.println(4.0 / 0.0); // Number > 0 / 0.0
> Infinity
System.out.println(-4.0 / 0.0); // Number < 0 / 0.0
> -Infinity
System.out.println(0.0 / 0.0); // 0.0 / 0.0 or number % 0.0
> NaN
```
Mixing ints and doubles
```
int a = 5;
double b = 3.2;
double c = a + b;
> 8.2
```
```
double a = 5.0;
double b = 3.5;
int c = (int)(a + b);
// If there's a double in an expression, all the other numbers will automatically be casted to a double (as they are encountered)
```
